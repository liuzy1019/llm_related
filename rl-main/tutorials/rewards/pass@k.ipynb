### passk@1 vs. pass@k

> p是正确率，pass@k和p是一一对应关系。 而且任何p>0的模型输出，pass@k 在k足够大都会接近1。

$$
1-(1-p)^k
$$

- pass@1
    - 奖励是独立的 (Independent): 对于一批采样出的 $n$ 个答案 $x_1, x_2, ..., x_n$，它们各自的原始奖励 $g(x_1), g(x_2), ..., g(x_n)$ 是独立计算的。$x_1$ 的奖励只跟 $x_1$ 本身有关，与其他样本无关。
    - 信息是局部的 (Local): 用于更新模型的奖励向量就是 $[g(x_1), g(x_2), ...]$。模型学到的信号是：“x_1 本身很好/很差”。
- pass@k 的计算 (以 PKPO 为例):
    - 奖励是相互依赖的 (Interdependent): 这是最本质的计算区别。要计算 $x_1$ 最终用于模型更新的“变换后奖励” $s_1$，必须知道所有其他样本的奖励 $g(x_2), ..., g(x_n)$。
    - 信息是全局的 (Global): 计算过程如网页解析中所述，需要：
        - 收集所有 n 个样本的原始奖励。
        - 对这 n 个奖励进行全局排序。
        - 应用一个复杂的变换函数（如 sloo_minus_one），这个函数会考虑所有样本的奖励值和它们的排序位置，来计算出每个样本对“集体目标”（即在 k 个样本中找到最大奖励）的实际贡献。
- 因此，用于更新模型的奖励向量是 [s_1, s_2, ..., s_n]，其中每一个 s_i 都是基于全局信息计算出来的。模型学到的信号是：“x_1 在当前这个样本集体中，是一个有价值的/没价值的尝试”。
