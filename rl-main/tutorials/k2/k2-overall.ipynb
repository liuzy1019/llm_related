### k2 vs. r1

https://huggingface.co/deepseek-ai/DeepSeek-R1/blob/main/modeling_deepseek.py
- config
    - num_attention_heads, num_key_value_heads
    - n_routed_experts, n_shared_experts, num_experts_per_tok
- r1
    - 模型包含 256 个路由专家（routed expert）和 1 个共享专家（shared expert）
    - 对于每个 token，路由机制会选择得分最高的 8 个专家进行激活
    - 当模型处理一个 token 时，会发生以下过程  ：
        - 首先，该 token 会经过共享专家进行处理。
        - 然后，一个路由机制会从 256 个路由专家中，为这个特定的 token 选择 8 个最合适的专家。
        - 最后，该 token 会继续被这 8 个选中的专家处理。
