### PG & bellman equation

- 策略梯度方法虽然直接优化策略，但其优化信号往往依赖于价值函数的估计（$V^\pi(s)$，$Q^\pi(s,a)$），而这些价值函数是通过贝尔曼方程定义或学习得到的。
- 马尔可夫决策过程（MDP）是强化学习的核心框架，用于建模代理与环境的交互，其中贝尔曼方程是评估策略质量和寻找最优策略的关键工具。
- 贝尔曼方程是强化学习中用来计算状态或状态-动作对长期回报的数学公式。
    - https://mp.weixin.qq.com/s/rgRJvP91IBUc0TQ0Z4wo2Q
    - 状态价值函数的贝尔曼方程：计算从某个状态开始遵循策略的期望总回报，比如在迷宫中，某个位置值多少钱。
        - $V_\pi(s)=\mathbb E_{\pi}[R_{t+1}+\gamma V_\pi(S_{t+1})|S_t=s]$
        - 展开为：$V_\pi(s)=\sum_a\pi(a|s)\sum_{s'}P(s'|s,a)[R(s,a,s')+\gamma V_\pi(s')]$
    - 动作价值函数的贝尔曼最优方程：计算从某个状态采取特定动作后遵循最优策略的期望回报，帮助选择最佳动作。
        - $Q_\pi(s,a)=\mathbb E_\pi[R_{t+1}+\gamma Q_\pi(S_{t+1},A_{t+1})|S_t=s,A_t=a]$
        - 展开为：$Q_\pi(s,a)=\sum_{s'}P(s'|s,a)[R(s,a,s')+\gamma \sum_{a'}\pi(a'|s')Q_\pi(s',a')]$
